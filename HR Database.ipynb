#### 55. Write a query to find the addresses (location_id, street_address, city, state_province, country_name) of all departments.

```python
%%sql
SELECT  d.department_name,
        l.location_id,
        l.street_address,
        l.city,
        l.state_province,
        c.country_name
FROM departments d
INNER JOIN locations l
    ON d.location_id = l.location_id
INNER JOIN countries c
    ON l.country_id = c.country_id;
```

```python
departments = pd.read_sql("SELECT * FROM departments", engine)
locations = pd.read_sql("SELECT * FROM locations", engine)
countries = pd.read_sql("SELECT * FROM countries", engine)
df_merged = pd.merge(departments, locations, on='location_id')
pd.merge(df_merged, countries, on='country_id')[['department_name', 'location_id', 'street_address', 'city', 'state_province', 'country_name']]
```

#### 59. Write a query to find the employee ID, job title and number of days an employee worked in department 90.

```python
%%sql
SELECT  j.employee_id,
        jt.job_title,
        end_date - start_date AS days
FROM job_history j
INNER JOIN jobs jt
    ON j.job_id = jt.job_id
WHERE j.department_id = 90;
```
